- Stacked bar graphs are ideal for displaying the composition of multiple categories across different groups or segments

```python
import pandas as pd

data = {
    'Age Group': ['18-24', '18-24', '18-24', '25-34', '25-34', '25-34', '35-44', '35-44', '35-44'],
    'Platform': ['Facebook', 'Instagram', 'Twitter', 'Facebook', 'Instagram', 'Twitter', 'Facebook', 'Instagram', 'Twitter'],
    'number of users': [150, 200, 100, 300, 250, 150, 100, 150, 50]
}

df = pd.DataFrame(data)
import matplotlib.pyplot as plt
import numpy as np

# Group the data by Age Group and Platform and calculate the sum of number of users
grouped_data = df.groupby(['Age Group', 'Platform'])['number of users'].sum().unstack()

# Get the unique platforms and age groups
platforms = df['Platform'].unique()
age_groups = df['Age Group'].unique()

# Create the stacked bar graph
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the stacked bars
for i, platform in enumerate(platforms):
    ax.bar(age_groups, grouped_data[platform], bottom=np.sum(grouped_data.values[:, :i], axis=1), label=platform)

# Add labels and title
ax.set_xlabel('Age Group')
ax.set_ylabel('Number of users')
ax.set_title('Social Media Platform Usage by Age Group')

# Show the legend and plot the graph
ax.legend()
plt.show()
```

- We can use scatter plot to identifying patterns, trends, and relationships between variables.

```python
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Create a scatter plot
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)

# Add labels and title
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')

# Show the plot
plt.show()
```

- We can also colour code our points to show different categories.

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')

# Create a scatter plot
sns.scatterplot(x='sepal_length', y='sepal_width',hue='species',data=iris)

# Add labels and title
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')

# Show the plot
plt.show()
```

- When it comes to visualizing the composition of different categories in a data set, two common graph types that come to mind are pie charts and bar graphs.

```python
iris['species'].value_counts().plot(kind='pie')
```

```python
iris['species'].value_counts().plot(kind='bar')
```

```python
sentiment_counts = iris['species'].value_counts()

# Create a stacked bar graph
plt.figure(figsize=(8, 6))
colors = ['blue','orange','green']  # Customize colors for each sentiment category

# Plotting the stacked bars
plt.bar(sentiment_counts.index, sentiment_counts, color=colors)

# Adding labels and title
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Distribution')

# Add legend

# Show the legend and plot the graph
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=iris, x='sepal_length', kde=True, hue='species')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=iris, x='species', y='sepal_length')
plt.title('Distribution of Sepal Length across Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(data=iris, x='species', y='sepal_length')
plt.title('Distribution of Sepal Length across Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```
